```python
print("Bem-vindo ao sistema de cadastro!") # Mensagem de boas-vindas

nome = input("Digite seu nome completo: ") # Solicita o nome do usuário  # exemplo de apenas texto
cpf = int (input("Digite seu CPF: ")) # Solicita o cpf do usuário  # exemplo de números inteiros
email = input("Digite seu e-mail: ") # Solicita o email  # exemplo de apenas texto
senha = input("Digite uma senha: ") # Normalmente, não mostramos a senha. # exemplo de apenas texto

print("\ncadastro realizado com sucesso!")
print("Nome:", nome)
print("CPF:", cpf)
print("E-mail:", email)
print("Senha:", senha)
```
